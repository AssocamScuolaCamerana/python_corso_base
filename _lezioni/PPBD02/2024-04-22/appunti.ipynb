### Estrazione dei codici dei caratteri da un font

Dentro il modulo custom `font_tools` ci sono due funzioni:

- `_get_font_path()` restituisce il percorso del file di un font a partire dal suo nome.
- `get_font_chars()` restituisce una tupla con tutti i caratteri non vuoti, e non emoji, contenuti in un font, dato il suo nome.

```python
from font_tools import _get_font_path

_get_font_path('arial.ttf')
```

```python
from font_tools import get_font_chars

get_font_chars('arial.ttf')
```
